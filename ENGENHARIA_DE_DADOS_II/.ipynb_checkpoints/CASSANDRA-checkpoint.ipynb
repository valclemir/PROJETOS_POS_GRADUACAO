```python
import pandas as pd 
import numpy as np
from cassandra.cluster import Cluster
from warnings import simplefilter

simplefilter('ignore')
```

#### Carrega o dataset

```python
df = pd.read_csv('dadoscovid.csv')
```

#### Seleciona colunas

```python
columns =['location', 'population', 'people_fully_vaccinated']
df = df[columns]
```

#### Mostra as informações do dataset

```python
df.info()
```

#### Tratamento dos dados

```python
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].fillna(0)
df['population'] = df['population'].astype('int64')
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].astype('int64')
df.loc[df['population'] < 0, 'population'] = [i.split('-')[1] for i in df['population'][df['population'] < 0].astype(str).values]
```

#### Abre a conexão com o cassandra

```python
cluster = Cluster(['bigdata-1'],port=9042)
session = cluster.connect()
```

#### Insere no Cassandra

```python
for i, linha in df.iterrows():
    id = i
    location = linha['location']
    population = linha['population']
    people_fully_vaccinated = linha['people_fully_vaccinated']
    
        
    SQL = f'''INSERT INTO bigdata.family_group (id, location, population, people_fully_vaccinated) 
            VALUES ({id}, '{location}', {population}, {people_fully_vaccinated});'''
    
    
    session.execute(SQL)
    
```

#### Carrega os dados do banco

```python
SQL = 'SELECT * FROM bigdata.FAMILY_GROUP'
df2= pd.DataFrame(list(session.execute(SQL)))
```

#### Fecha a conexão 

```python
session.shutdown() 
```

```python
paises = df[columns]
```

#### Trata os dados negativos

```python
paises.loc[paises['population'] < 0, 'population'] = [i.split('-')[1] for i in paises['population'][paises['population'] < 0].astype(str).values]
```

#### Agrupa os dados

```python
total_vacinados = paises[['location', 'people_fully_vaccinated']].groupby('location').max().reset_index()['people_fully_vaccinated'].values
```

#### Dropa dados duplicados para formar um novo dataframe

```python
new_df = paises[['location', 'population']].drop_duplicates()
new_df['total_vacinados'] = total_vacinados
new_df.head(4)
```

#### Calcula o percentual de vacinados por paises e ordena o dataset da maior taxa para menor

```python
new_df['taxa'] = new_df['total_vacinados'] / df['population'] * 100
new_df = new_df.sort_values('taxa', ascending=True)
```

```python
new_df.head(10)
```

#### Plota grafico de barras com os percentuais

```python
import plotly.graph_objects as go
layout = go.Layout(yaxis=dict(tickformat=".2%"))
fig = go.Figure(go.Bar(
            x=new_df['taxa'],
            y=new_df['location'],
            orientation='h'),
            layout = layout)

fig.show()
```
