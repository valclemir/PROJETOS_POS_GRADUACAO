```python
import pandas as pd 
import numpy as np 
import requests as res
from warnings import simplefilter
import seaborn as sns

simplefilter("ignore")
```

```python
response = res.get("https://apisidra.ibge.gov.br/values/t/5457/n6/2300903,2306306,2306405,2310209,2310704,2312403,2312601,2313500,2313807,2302206,2303501,2302800,2303006,2304608,2310407,2301208,2301406,2302107,2302909,2305100,2306504,2309102,2309805,2310100,2311603,2306603,2300754,2310258,2313351,2313559,2313757,2303956,2310852,2300150,2301950,2309458/v/215,8331/p/last%201/c782/all").json()
```

```python
columns = []
for i in pd.DataFrame(response).iloc[0]:
    columns.append(i)
```

```python
df = pd.DataFrame(response).iloc[1:, :]
```

```python
df.columns = columns
```

```python
df.head()
```

```python
df[df["Município"].str.contains("(CE)")].head()
```

```python
df.columns
```

```python
df = df[df["Município"].str.contains("CE")]
```

```python
df["Valor"] = df["Valor"].replace("-", 0).replace("...", 0).astype(float)
```

```python
columns_select = ["Município", "Ano", "Valor", ]
columns_groupby = ["Município", "Ano",]
```

```python
custo_total_por_municipio_e_ano = df[columns_select].groupby(columns_groupby).sum().reset_index().sort_values("Valor", ascending=False)
```

```python
media = custo_total_por_municipio_e_ano["Valor"].mean()
```

```python
desvio_padrao = custo_total_por_municipio_e_ano["Valor"].std()
```

```python
custo_total_por_municipio_e_ano["Z-SCORE"] = (custo_total_por_municipio_e_ano["Valor"] - media) / desvio_padrao
custo_total_por_municipio_e_ano["% COEF_VAR"] = round((desvio_padrao / media) * 100, 2)
custo_total_por_municipio_e_ano["% DISTANCIA_DA_MEDIA "] = round((1 - (desvio_padrao / media)) * 100, 2)
```

```python
custo_total_por_municipio_e_ano
```

```python
columns_select = ["Município", "Ano", "Valor","Produto das lavouras temporárias e permanentes"]
columns_groupby = ["Município", "Ano", "Produto das lavouras temporárias e permanentes"]
custo_total_por_municipio_ano_e_produtos = df[columns_select].groupby(columns_groupby).sum().reset_index()
```

```python
custo_total_por_municipio_ano_e_produtos.head()
```

```python
media = custo_total_por_municipio_ano_e_produtos["Valor"].mean()
desvio_padrao = custo_total_por_municipio_ano_e_produtos["Valor"].mean()
```

```python
custo_total_por_municipio_ano_e_produtos["Z-SCORE"] = (custo_total_por_municipio_ano_e_produtos["Valor"] - media) / desvio_padrao
custo_total_por_municipio_ano_e_produtos["% COEF_VAR"] = round( (desvio_padrao / media * 100) , 2)
custo_total_por_municipio_ano_e_produtos["% DISTANCIA_DA_MEDIA"] = round((1 - (desvio_padrao / media)) * 100, 2)
```
