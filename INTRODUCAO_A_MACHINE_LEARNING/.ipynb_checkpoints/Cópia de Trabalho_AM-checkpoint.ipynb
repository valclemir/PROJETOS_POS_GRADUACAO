```python
#Tratando Colunas
df_salario.columns = [i.strip() for i in df_salario.columns.tolist()]
df_salario['AGE'] = df_salario['AGE'].str.replace('-', '0')
df_salario['AGE'] = df_salario['AGE'].astype(int)
df_salario['SALARY'] = df_salario['SALARY'].str.replace('Not signed','0')
df_salario['SALARY'] = df_salario['SALARY'].str.replace(',', '')
df_salario['SALARY'] = df_salario['SALARY'].astype('float')
```

#Conjunto de Treinamento e teste


###Amostragem Aleatória

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
test_size=0.2 # Test set size: 20%
random_state=42 # Definir numero aleatorio

scalerx = StandardScaler().fit(df_x_salario)
scalery = StandardScaler().fit(df_y_salario)

X_scaler = scalerx.fit_transform(df_x_salario)
Y_scaler = np.squeeze(scalery.fit_transform(df_y_salario))

x_train, x_test, y_train, y_test = train_test_split(X_scaler, Y_scaler,
                                      test_size=test_size,
                                      random_state=1) # Use if data is large enough


#Modelo 1
LR = LinearRegression()
LR.fit(x_train, y_train)
LR.score(x_train, y_train)

#Modelo 2 
LRSGDRegressor = SGDRegressor(loss="squared_loss", penalty="l2", max_iter=1000).fit(x_train, y_train)  
LRSGDRegressor.score(x_train, y_train)


#Modelo 3
dec_tree_reg=DecisionTreeRegressor().fit(x_train, y_train) 
dec_tree_reg.score(x_train,y_train)

#Modelo 4
rand_forest_reg=RandomForestRegressor()
rand_forest_reg.fit(x_train, y_train)

y_pred = LR.predict(x_test)
print('Modelo 1')
print(f'Score :{LR.score(x_train, y_train)}')
print(f'Erro médio quadrado : {mean_squared_error(y_test, y_pred)}')                                      
print(f'Erro médio absoluto : {mean_absolute_error(y_test, y_pred)}')
print(f'Raiz quadrada erro médio: {np.sqrt(mean_squared_error(y_test, y_pred))}')

ax1 = sns.distplot(y_test, hist=False, color="r", label="Valor atual")
plot = sns.distplot(y_pred, hist=False, color="b", label="Valores preditos" , ax=ax1)
plot


y_pred = scalery.inverse_transform(y_pred)
df_final = pd.DataFrame()
df_final['atual'] = scalery.inverse_transform(y_test)
df_final['predito'] = np.around(y_pred , 2)
```

```python
#agrupando por "POSITION"
df_group_pos = df_salario.groupby('POSITION')
#gera arquivo com os dados agupados por 'POSITION'
dados = df_group_pos['SALARY'].mean().reset_index().sort_values('SALARY',ascending=False)
# cria coluna com % de Salario
dados['% Sal_Pos'] = round(dados['SALARY']/dados['SALARY'].sum(),3)*100
```

```python
#gera o gráfico
#define tamanho do plot
fig,axes=plt.subplots(figsize=(9,7))
#gera o grafico de barras
plt.style.use('seaborn')
ax,texto,autotexto = axes.pie(dados['% Sal_Pos'],wedgeprops=dict(width=0.5),autopct='%1.1f%%',startangle=90,labels= dados['POSITION']);
axes.set_title('% de Salários por Posição',fontsize=20)
plt.setp(autotexto, size=10, weight="bold");
for t in texto:
    t.set_size('15');

plt.show();
```

Em geral, existe uma grande diferença entre as posições Guarda de Tiro e C e o Ala-Pivô, nas demais posições existe diferenças minimas de salario.
