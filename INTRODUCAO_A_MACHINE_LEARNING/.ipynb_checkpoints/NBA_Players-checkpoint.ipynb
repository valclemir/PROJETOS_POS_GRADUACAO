## 1.1 Obtendos dados/ limpeza e transformação

```python
import pandas as pd
import numpy as np  
```

```python
# Carregando o arquivo
df = pd.read_csv("NBA_Players.csv",sep=',',encoding='utf-8')
```

```python
# copia do dataset
df_nba = df.copy()
```

```python
df_nba.info()
```

```python
df_nba.describe().T
```

```python
df_nba.head(5).T
```

```python
# Retirar os espaços das colunas
df_nba.columns = df_nba.columns.str.replace(' ', '') 
```

```python
df_nba.columns
```

```python
# separar as colunas FGM_FGA, THM_THA, FTM_FTA, pois consta duas informações na mesma coluna
df_nba = df_nba.join(df_nba['FGM_FGA'].str.split('-', expand=True).rename(columns={0:'FGM', 1:'FGA'}))
df_nba = df_nba.join(df_nba['THM_THA'].str.split('-', expand=True).rename(columns={0:'THM', 1:'THA'}))
df_nba = df_nba.join(df_nba['FTM_FTA'].str.split('-', expand=True).rename(columns={0:'FTM', 1:'FTA'}))
```

```python
# excluido colunas depois de separadas.
df_nba.drop(['FGM_FGA','THM_THA','FTM_FTA'], axis=1, inplace=True)
```

```python
df_nba.head(5).T
```

```python
# substituindo "Not Signed" da coluna SALARY por 0
df_nba.replace("Not signed",0, inplace=True)
```

```python
# retirar a virgula do salario e converter para float e verificando as 10 primeiras alterações
df_nba['SALARY'] = df_nba['SALARY'].str.replace(',','').astype(float)
```

```python
# Transformando a notação cientifica
pd.options.display.float_format = '{:.1f}'.format
```

```python
df_nba.head(5).T
```

```python
# removendo linhas com numeros negativos da coluna PER_LAST_SEASON
df_nba.drop(df_nba[df_nba.PER_LAST_SEASON < 0].index, inplace=True)
```

```python
df_nba.info()
```

```python
#Verificando numeros ausentes 
df_nba.isnull().sum()
```

```python
# preenche com 0 os values null das colunas PPG_LAST_SEASON, APG_LAST_SEASON, RPG_LAST_SEASON,PER_LAST_SEASON 
df_nba['APG_LAST_SEASON'].fillna(0, inplace=True)
df_nba['PPG_LAST_SEASON'].fillna(0, inplace=True)
df_nba['RPG_LAST_SEASON'].fillna(0, inplace=True)
df_nba['PER_LAST_SEASON'].fillna(0, inplace=True)
df_nba['SALARY'].fillna(0, inplace=True)
df_nba['FTA'].fillna(0, inplace=True)
df_nba['THA'].fillna(0, inplace=True)
df_nba['FGA'].fillna(0, inplace=True)
```

```python
df_nba['AGE'].value_counts()
```

```python
# Remover '-' da coluna AGE
df_nba.drop(df_nba[df_nba.AGE == '-'].index, inplace=True)
```

```python
# Converter colunas para float
df_nba['AGE'] = df_nba['AGE'].astype(float)
df_nba['EXPERIENCE'] = df_nba['EXPERIENCE'].astype(float)
df_nba['GP'] = df_nba['GP'].astype(float)
df_nba['FTA'] = df_nba['FTA'].astype(float)
df_nba['THA'] = df_nba['THA'].astype(float)
df_nba['FGA'] = df_nba['FGA'].astype(float)
df_nba['FGM'] = df_nba['FGM'].astype(float)
df_nba['THM'] = df_nba['THM'].astype(float)
df_nba['FTM'] = df_nba['FTM'].astype(float)
```

```python
# verificando as categorias de POSITION
df_nba['POSITION'].value_counts()
```

```python
# Removendo nomes duplicados se houverem
df_nba.drop_duplicates(['NAME'], inplace=True)
```

```python
#Excluindo colunas que não usaremos
df_nba.drop([], axis=1, inplace=True)
```

```python
# Fazendo copoia do dataset limpo
nba = df_nba.copy()
```

#### Tratando valores misiing ocultos na coluna salario

### Verificando a Correlação entre as variáveis

```python
# Importando os módulos    
import matplotlib as mat
import matplotlib.pyplot as plt    
import sklearn as sk
%matplotlib inline  
```

```python
# Função para Correlção entre as varíaveis
def plot_corr(df, size=10):
    corr = df.corr()    
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr) 
    ax = plt.gca() # get the current axes
    ax.set_xticks(range(len(corr.columns)))
    ax.set_yticks(range(len(corr.columns)))
    ax.set_xticklabels(corr.columns)
    ax.set_yticklabels(corr.columns)
    plt.setp(ax.get_xticklabels(), rotation=45, ha="left", rotation_mode="anchor")
```

```python
# Criando o gráfico
plot_corr(nba)
```

```python
corr_nba = nba.corr()
```

```python
# Visualizando a correlação em tabela
# Coeficiente de correlação: 
# +1  = forte correlação positiva
# 0   = não há correlação
# -1  = forte correlação negativa
corr_nba['SALARY'].sort_values(ascending=False) # Ordenando a coluna, variavel alvo
```

## Pipeline para atributos numéricos e categóricos

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer



nba_attribs=['PPG_LAST_SEASON','PPG_CAREER','PPG',
             'FTM','FGM']
            # 'FTA','FGA','MPG','TOPG',
             #'STLPG','APG_LAST_SEASON',
             #'GP','EXPERIENCE']

nba_cat=['POSITION']

nba_pipeline=Pipeline([('imputer', SimpleImputer(missing_values = 0, strategy = "mean")),
                       ('scaler', StandardScaler())])

nba_cat_pipeline=Pipeline([('imputer', SimpleImputer(strategy='constant',fill_value='missing')),
                         ('encoder', OneHotEncoder(sparse=False))])

nba_full_pipeline=ColumnTransformer(transformers=[('nba_pipeline', nba_pipeline, nba_attribs),
                                              ('nba_cat_pipeline', nba_cat_pipeline, nba_cat)])

```

## Conjunto de Validação de Dados - Treino 60% , Validação 20% e Teste 20%

```python
from sklearn.model_selection import train_test_split

test_size=0.2 
random_state=42 
strat_train_set0,strat_test_set0=train_test_split(nba,
                                                  test_size=test_size,
                                                  random_state=random_state)
                                    

valid_size=0.25 
random_state=24 
strat_train_set0,strat_valid_set0=train_test_split(strat_train_set0,
                                                   test_size=valid_size,
                                                   random_state=random_state)


print(strat_train_set0.info())
print(strat_valid_set0.info())
print(strat_test_set0.info())
```

### Definindo variaveis para o treino

```python
nba_treino = strat_train_set0.drop('SALARY',axis=1) # X
nba_labels_treino = strat_train_set0['SALARY'].copy() # y
nba_treino_prepared = nba_full_pipeline.fit_transform(nba_treino) # X
```

### Definindo variaveis para a validação

```python
nba_val = strat_valid_set0.drop('SALARY',axis=1) # X
nba_labels_val = strat_valid_set0['SALARY'].copy() # y

nba_val_prepared = nba_full_pipeline.fit_transform(nba_val) # X
```

### Treinando modelo Decision Treee Regressor

```python
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor()
dec_tree_reg.fit(nba_treino_prepared,nba_labels_treino)
```

```python
nba_predicted=dec_tree_reg.predict(nba_val_prepared)
dec_tree_rmse=np.sqrt(mean_squared_error(nba_labels_val,nba_predicted))

print(dec_tree_rmse)
```

### Regressão Linear

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lin_reg=LinearRegression()
scores=cross_val_score(lin_reg,
                       nba_treino_prepared,
                       nba_labels_treino,
                       scoring='neg_mean_squared_error', # cross_val_score expects a utility function
                       cv=4) # quantas iterações tera o k-fold

lin_reg_rmse_score=np.sqrt(-scores) # cross_val_score expects a utility function

print(lin_reg_rmse_score)
print(lin_reg_rmse_score.mean())
print(lin_reg_rmse_score.std())
```

### Árvore de Decisão

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

dec_tree_reg=DecisionTreeRegressor()

scores=cross_val_score(dec_tree_reg,
                       nba_treino_prepared,
                       nba_labels_treino,
                       scoring='neg_mean_squared_error',
                       cv=4)

dec_tree_reg_rmse_score=np.sqrt(-scores)

print(dec_tree_reg_rmse_score)
print(dec_tree_reg_rmse_score.mean())
print(dec_tree_reg_rmse_score.std())
```

#### Grid Search

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rand_forest_reg = RandomForestRegressor()

param_grid=[{'n_estimators':[3,10,30],'max_features':[2,4,6,8]}, # 3x4=12
            {'bootstrap':[False],'n_estimators':[3,10],'max_features':[2,3,4]}] # 1x2x3=6

grid_search=GridSearchCV(rand_forest_reg,
                         param_grid,
                         scoring='neg_mean_squared_error',
                         refit=True,
                         cv=5, #5
                         return_train_score=True) 

grid_search.fit(nba_treino_prepared,nba_labels_treino) 

print(grid_search.best_params_)
```

### Randomized Search

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

rand_forest_reg = RandomForestRegressor()

param_grid=[{'n_estimators': np.arange(100), 'max_features': np.arange(1,10)}, 
            {'bootstrap':[False],'n_estimators': np.arange(50), 'max_features':np.arange(1,10)}] 

randomized_search=RandomizedSearchCV(rand_forest_reg,
                         param_grid,
                         n_iter=10, 
                         scoring='neg_mean_squared_error',
                         refit=True,
                         cv=10, #10
                         random_state=42, # Seed
                         return_train_score=True) 

randomized_search.fit(nba_treino_prepared,nba_labels_treino)

print(randomized_search.best_params_)
```

```python
# Definindo conjunto de teste
nba_teste = strat_test_set0.drop('SALARY',axis=1) # X
nba_labels_teste = strat_test_set0['SALARY'].copy() # y
nba_teste_prepared = nba_full_pipeline.fit_transform(nba_teste) # X
```

### Analise no conjunto de teste

```python
d=randomized_search.cv_results_

for mean_train_score,mean_test_score,params in zip(d['mean_train_score'],d['mean_test_score'],d['params']):
    print(np.sqrt(-mean_train_score),np.sqrt(-mean_test_score),params)
```

```python
from sklearn.metrics import mean_squared_error


rand_forest_reg = randomized_search.best_estimator_ 
teste_predicted = rand_forest_reg.predict(nba_teste_prepared)
rand_forest_reg_rmse = np.sqrt(mean_squared_error(nba_labels_teste,teste_predicted))

print(rand_forest_reg_rmse)
```
