```python
import os 
import pandas as pd
import pandas as pd 
```

```python
path = "datasets\datasets\genius"
full_text = []


for root, dirs, files in os.walk(path, topdown=True):
    for file in files:
        dict_season = {}
        
        if file != "ref.txt":
            dir_full = os.path.join(root, file)
            #file_r = open(dir_full, encoding="utf8")
            with open(dir_full, encoding="utf8") as file_r:
                file_r = file_r.read()
            dict_season = {"season": file, "text": file_r}
            full_text.append(dict_season)
```

```python
full_text[0]["text"]
```

```python
text = full_text[0]["text"].split("scene shifts")
```

```python
list_person = []
df_person = pd.DataFrame()
for i in range(len(text)):
    print(f"""----------CONVERSA {i}---------- """)
    
    

    for j in text[i].split("\n"):
        dict_ = {}
        if j.find(".:"):
            
            t = j[0:j.find(":")]
            if len(t) < 20:
                if t == "WAYMAR ROYCE":
                    person =  t.split(" ")[1].replace("[", "")
                
                else:
                    person =  t.split(" ")[0].replace("[", "")
                
                
                #list_person.append(dict_)
                if person.split(" ")[0] != "" and person.split(" ")[0] != "TITLE" \
                and  person.split(" ")[0] != "Blackout" and person.split(" ")[0] != "/":
                    print( person)
                    dict_["conversa"] = i
                    dict_["person"] = person
                    list_person.append(dict_)
    
    
    
    df = pd.DataFrame(list_person)[["person"]]
    df1 = pd.DataFrame(list_person)[["person"]]

    df['key']  = 1
    df1['key'] = 1
    
    
    result = pd.merge(df, df1, on ='key').drop("key", 1)
    result["conversa"] = i
    df_person = df_person.append(result)
```

```python
dd = df_person.drop_duplicates()[df_person.drop_duplicates()["conversa"] == 1]
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.from_pandas_edgelist(dd, 'person_x', 'person_y')
```

```python
pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edge_color='r', arrows = True)

plt.show()
```
